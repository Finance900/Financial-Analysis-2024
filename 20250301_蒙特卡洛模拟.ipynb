```python
#计算宁德时代和贵州茅台协方差并计算相关性
```

```python
#对齐宁德时代和贵州茅台日期索引
df1,df2=df1.align(df2,join='inner')

#计算日算收益率
returns_maotai=df1['close'].pct_change().dropna()
returns_ningde=df2['close'].pct_change().dropna()

#协方差矩阵
cov_matrix =np.cov(returns_maotai ,returns_ningde)
print('协方差矩阵：\n',cov_matrix)

#相关系数
corr_coef =np.corrcoef(returns_maotai ,returns_ningde)[0,1]
print(f'相关系数：{corr_coef:.2f}')
```

```python
#验证分散投资组合能否降低风险
```

```python
# 组合波动率公式：σ_p = sqrt(w1²σ1² + w2²σ2² + 2w1w2Cov(A,B))
weights =np.array([0.6,0.4])
portfolio_vol =np.sqrt(weights @ cov_matrix @ weights.T)
print(f'组合年化波动率：{portfolio_vol *np.sqrt(242):.2%}')

#对比等权组合
weights_equal =np.array([0.5,0.5])
vol_equal =np.sqrt(weights_equal @ cov_matrix @weights_equal.T)
print(f'等权组合波动率：{vol_equal *np.sqrt(242):.2%}')
```

```python
weights_list = [np.array([w, 1-w]) for w in np.linspace(0, 1, 50)]  # 权重从0%~100%
volatility_list = [np.sqrt(w @ cov_matrix @ w.T) * np.sqrt(252) for w in weights_list]

# 绘制有效前沿
plt.plot([w[0] for w in weights_list], volatility_list, label='组合波动率')
plt.scatter(0.6, portfolio_vol * np.sqrt(242), color='red', label='60%茅台组合')
plt.xlabel("茅台权重")
plt.ylabel("年化波动率")
plt.title("投资组合波动率 vs 权重分配")
plt.legend()
plt.savefig("组合波动率曲线.png")
plt.show()
```

```python
#随着茅台权重的增加，组合波动率逐渐下降，这是因为茅台波动率远低于宁德时代波动率。
```

```python
import numpy as np

# 计算茅台和宁德时代的收益率和风险（日收益率序列，如2月26日数据）
annual_return1 = np.mean(returns_maotai) * 242         # 年化平均收益
annual_volatility1 = np.std(returns_maotai) * np.sqrt(241)  # 年化波动率（已会）
annual_return2 =np.mean(returns_ningde)*242
annual_volatility2 =np.std(returns_ningde)* np.sqrt(242)

#计算夏普比率
risk_free_rate = 0.025                                # 无风险利率2.5%
maotai_sharpe_ratio = (annual_return1 - risk_free_rate) / annual_volatility1
ningde_sharpe_ratio =(annual_return2 -risk_free_rate)/annual_volatility2

print(f"茅台夏普比率：{maotai_sharpe_ratio:.2f}")
print(f'宁德时代夏普比率：{ningde_sharpe_ratio:.2f}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. 生成随机点：在正方形 [-1,1] x [-1,1] 内撒 10000 个点
n_points = 10000
x = np.random.uniform(-1, 1, n_points)  # 生成 x 坐标
y = np.random.uniform(-1, 1, n_points)  # 生成 y 坐标

# 2. 判断点是否在圆内：到原点的距离 <=1
inside_circle = (x**2 + y**2) <= 1

# 3. 计算落在圆内的比例
pi_estimate = 4 * np.sum(inside_circle) / n_points

# 4. 输出结果
print(f"估算的 π 值: {pi_estimate}")
print(f"真实 π 值: {np.pi}")

# 5. 可视化（可选）
plt.scatter(x[inside_circle], y[inside_circle], color='blue', s=1)
plt.scatter(x[~inside_circle], y[~inside_circle], color='red', s=1)
plt.axis('equal')
plt.show()
```
