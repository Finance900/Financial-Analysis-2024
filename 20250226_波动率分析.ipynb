```python
import tushare as ts
import pandas as pd

#设置token
ts.set_token('your token')
pro=ts.pro_api()

#获取贵州茅台2023年日线数据，贵州茅台代码为600519.SH
df1=pro.daily(ts_code='600519.SH',start_date='20240101',end_date='20241231')

#保存为excel
df1.to_excel('茅台_2024.xlsx')
print('数据已保存')
```

```python
df1
```

```python
import matplotlib.pyplot as plt

#转换数据日期格式
df1['trade_date']=pd.to_datetime(df1['trade_date'],format='%Y%m%d')

# 设置字体为黑体
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

#绘制收盘价曲线
plt.figure(figsize=(10,6))
plt.plot(df1['trade_date'],df1['close'],label='贵州茅台收盘价',color='red',linewidth=2)
plt.xlabel('日期')
plt.ylabel('价格（元）')
plt.title('2024贵州茅台收盘价走势')
plt.legend()
plt.grid(True)
plt.savefig('gui茅台收盘价.png')   #保存图片
plt.show()
```

```python
import numpy as np

#计算日收益率（今日收盘价/昨日收盘价-1）
df1['daily_return']=df1['close'].pct_change()

#删除缺失值（第一天无收益）
df1=df1.dropna()

#计算年化波动率
daily_volatility = np.std(df1['daily_return'])
annualized_volatility =daily_volatility *np.sqrt(252)

print(f'贵州茅台每天的波动率：{daily_volatility:.2%}')
print(f'贵州茅台2024年年化波动率：{annualized_volatility:.2%}')
```
