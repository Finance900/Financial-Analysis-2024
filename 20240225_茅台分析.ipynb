```python


import tushare as ts
import pandas as pd

#设置token
ts.set_token('your_token')
pro=ts.pro_api()

#获取贵州茅台2023年日线数据，贵州茅台代码为600519.SH
df=pro.daily(ts_code='600519.SH',start_date='20240101',end_date='20241231')

#保存为excel
df.to_excel('茅台_2024.xlsx')
print('数据已保存')
```

```python
import matplotlib.pyplot as plt

#转换数据日期格式
df['trade_date']=pd.to_datetime(df['trade_date'],format='%Y%m%d')

# 设置字体为黑体
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

#绘制收盘价曲线
plt.figure(figsize=(10,6))
plt.plot(df['trade_date'],df['close'],label='贵州茅台收盘价',color='red',linewidth=2)
plt.xlabel('日期')
plt.ylabel('价格（元）')
plt.title('2024贵州茅台收盘价走势')
plt.legend()
plt.grid(True)
plt.savefig('gui茅台收盘价.png')   #保存图片
plt.show()
```
